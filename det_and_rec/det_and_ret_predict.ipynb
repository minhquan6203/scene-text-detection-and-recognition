```python
%cd /content
import numpy as np
import json
j={}
file_list=sorted(os.listdir('/content/output'))
# file_list_img=sorted(os.listdir('/content/crop_img_test/'))
for file_ in file_list:
  with open('/content/output/'+file_) as f:
   lines=f.readlines()
   for i in range(len(lines)):
      lines[i]=lines[i].split(',')
      j[f"{file_.replace('.txt',f'_{i}_train.jpg')}"]=lines[i][8].replace('\n','')

out_file = open("train_crop.json", "w")
json.dump(j, out_file, indent = 4, sort_keys = False,ensure_ascii=False)
out_file.close()
```

```python
with open("/content/train_crop.json") as f:
  data=json.load(f)
i=0
for file_ in os.listdir('/content/crop_img_train'):
  if file_ not in data.keys():
    try:
      os.remove('/content/crop_img_train/'+file_)
    except Exception:
      pass
```

```python
d=[]
ans=[]
filenames=[]
ex=""
ques=""
delete=True
filepath="non-text-QA-images"
for k,v in data.items():
  an={"answer":v,"explaination":ex,"question":ques}
  d.append({
      "annotations":[an],
      "delete":delete,
      "filename":k,
      "filepath":filepath,
      
  })

out_file = open("train_crop_label.json", "w")
json.dump(d, out_file, indent = 4, sort_keys = False,ensure_ascii=False)
out_file.close()
```
